```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/drive/MyDrive/Machine Learning/DataSets/canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income(US$)')
plt.scatter(df.year, df['per capita income (US$)'], color = 'red', marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df[['per capita income (US$)']])
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
